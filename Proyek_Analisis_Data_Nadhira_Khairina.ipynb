### Question 2 : what was the purchasing frequency and monetary based on order?

RFM analysis is one of the commonly used methods for customer segmentation (grouping customers into several categories) based on three parameters: recency, frequency, and monetary. However, for data adaptation, segmentation based on orders is conducted with the following criteria:

Recency: This parameter is based on the shipping limit date.

Frequency: This parameter is used to identify how often a customer makes transactions.

Monetary: This parameter is used to identify the revenue generated based on the order price.

Based on these three parameters, we can identify which sales items have high value (frequent transactions and generating significant revenue) and low value.

```python
# RFM Analysis
# Grouping based on order_id and calculating the required attributes
rfm_df = sales_df.groupby(by="order_id", as_index=False).agg({
    "order_item_id": "nunique", # Counting the number of orders
    "price": "sum", # Calculating the total price
    "shipping_limit_date": "max" # Taking the latest shipping date
})

# Renaming columns for better understanding
rfm_df.columns = ["order_id", "frequency", "monetary", "max_order_item_id"]

# Calculating recency (number of days since the last shipping date)
recent_date = pd.to_datetime(rfm_df["max_order_item_id"]).max().date()
rfm_df["recency"] = rfm_df["max_order_item_id"].apply(lambda x: (recent_date - x.date()).days)

# Dropping the max_order_date column as it is no longer needed
rfm_df.drop("max_order_item_id", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

The RFM statistics indicate that, on average, there is 1 order with an average purchase value of 137.75. Generally, the shipping limit has an average range of about 823 days since the last transaction. The high standard deviation in the monetary attribute suggests significant variation in customer purchase orders. The median recency of 804 days indicates that most orders occur within a relatively short time frame of the shipping limit.







```python
# Sort order based on recency, frequency, and monetary score
rfm_df["r_rank"] = rfm_df["recency"].rank(ascending=False)
rfm_df["f_rank"] = rfm_df["frequency"].rank(ascending=True)
rfm_df["m_rank"] = rfm_df["monetary"].rank(ascending=True)
```

```python
# Normalize the rank of the customers based on recency, frequency, and monetary scores
rfm_df["r_rank_norm"] = (rfm_df["r_rank"] / rfm_df["r_rank"].max()) * 100
rfm_df["f_rank_norm"] = (rfm_df["f_rank"] / rfm_df["f_rank"].max()) * 100
rfm_df["m_rank_norm"] = (rfm_df["m_rank"] / rfm_df["m_rank"].max()) * 100
# Drop the original rank columns
rfm_df.drop(["r_rank", "f_rank", "m_rank"], axis=1, inplace=True)
```

```python
# Calculate RFM Score
rfm_df["RFM_score"] = (
    0.15 * rfm_df["r_rank_norm"]
    + 0.28 * rfm_df["f_rank_norm"]
    + 0.57 * rfm_df["m_rank_norm"]
)

# Scale the RFM Score
rfm_df["RFM_score"] *= 0.05

# Round the RFM Score to 2 decimal places
rfm_df = rfm_df.round(2)

# Display the top 5 customers based on RFM Score
rfm_df[["order_id", "RFM_score"]].head(5).sort_values(
    by="RFM_score", ascending=False
)
```

The higher the RFM score, the higher the level of activity and transaction value of the customers. In this example, the order with the first order ID has the highest RFM score, indicating that the order has a good combination of high recency, frequency, and monetary value. Conversely, the order with the last order ID has the lowest RFM score, indicating lower levels of activity and transaction value. By considering the RFM scores, businesses can identify more valuable customers and direct marketing strategies and services more effectively.

```python
# Create subplots with a specified size
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

# Define colors for the plots (purple)
colors = ["#8a2be2", "#8a2be2", "#8a2be2", "#8a2be2", "#8a2be2"]

# Plot for Recency
sns.barplot(y="recency", x="order_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel("Recency (days)", fontsize=15)
ax[0].set_xlabel(None)
ax[0].set_title("Top 5 by Recency", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=12)
ax[0].tick_params(axis='y', labelsize=12)

# Plot for Frequency
sns.barplot(y="frequency", x="order_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel("Frequency", fontsize=15)
ax[1].set_xlabel(None)
ax[1].set_title("Top 5 by Frequency", fontsize=18)
ax[1].tick_params(axis='x', labelsize=12)
ax[1].tick_params(axis='y', labelsize=12)

# Plot for Monetary
sns.barplot(y="monetary", x="order_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel("Monetary", fontsize=15)
ax[2].set_xlabel(None)
ax[2].set_title("Top 5 by Monetary", fontsize=18)
ax[2].tick_params(axis='x', labelsize=12)
ax[2].tick_params(axis='y', labelsize=12)

# Set x-axis labels as a list below the plot for all subplots
x_labels_recency = rfm_df.sort_values(by="recency", ascending=True).head(5)["order_id"].tolist()
x_labels_frequency = rfm_df.sort_values(by="frequency", ascending=False).head(5)["order_id"].tolist()
x_labels_monetary = rfm_df.sort_values(by="monetary", ascending=False).head(5)["order_id"].tolist()

ax[0].set_xticklabels(x_labels_recency, rotation=45, ha='right')
ax[1].set_xticklabels(x_labels_frequency, rotation=45, ha='right')
ax[2].set_xticklabels(x_labels_monetary, rotation=45, ha='right')

# Add a title for the overall visualization
plt.suptitle("Top Customers Based on RFM Parameters (Order ID)", fontsize=20)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```
